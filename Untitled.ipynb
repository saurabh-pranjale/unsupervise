```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
```

```python
df = pd.read_csv("customer_segmentation_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
X = df[['Annual_Income', 'Spending_Score']]
```

```python
plt.figure(figsize=(7,6))

plt.scatter(
    X['Annual_Income'],
    X['Spending_Score']
)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Distribution")

plt.show()
```

```python
#Find the Best Number of Clusters (Elbow Method)
```

```python
wcss = []

for i in range(1,11):

    model = KMeans(
        n_clusters=i,
        random_state=42
    )

    model.fit(X)

    wcss.append(model.inertia_)
```

```python
plt.figure(figsize=(7,5))

plt.plot(
    range(1,11),
    wcss,
    marker='o'
)

plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.title("Elbow Method")

plt.show()
```

```python
#Train the Final K-Means Model
```

```python
model = KMeans(
    n_clusters=5,
    random_state=42
)

model.fit(X)
```

```python
clusters = model.predict(X)
```

```python
# Add Cluster Column
```

```python
df['Cluster'] = clusters
```

```python
df.head()
```

```python
model.cluster_centers_
```

```python
plt.figure(figsize=(8,6))

plt.scatter(
    X['Annual_Income'],
    X['Spending_Score'],
    c=df['Cluster']
)

plt.scatter(
    model.cluster_centers_[:,0],
    model.cluster_centers_[:,1],
    marker='X',
    s=300
)

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Customer Segments")

plt.show()
```
